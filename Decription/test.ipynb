```python
import numpy as np
from matplotlib import pyplot as plt

# Load the key and the image
key = np.load('key.npy')
img = plt.imread('secret.bmp')

# Print shapes of image and key
print(f"Shape of the image: {img.shape}")
print(f"Shape of the key: {key.shape}")

# Check if the key's dimension matches the image's width
if key.shape[0] != img.shape[1]:
    raise ValueError("The key's dimension does not match the image's width.")

# Expand the key to match the image dimensions
# For RGB images, the key needs to be expanded to match the number of color channels
# key_expanded should have the shape (height, width) but need to match (height, width, channels)
key_expanded = np.tile(key, (img.shape[0], 1))  # Expand the key along the height
key_expanded_rgb = np.expand_dims(key_expanded, axis=-1)  # Add a channel dimension
key_expanded_rgb = np.tile(key_expanded_rgb, (1, 1, img.shape[2]))  # Repeat the key across all color channels

# Decrypt the image by applying XOR operation
decrypted_img = img ^ key_expanded_rgb

# Display the decrypted image
plt.imshow(decrypted_img)
plt.title('Decrypted Image')
plt.axis('off')  # Turn off axis labels
plt.show()
```
